```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

SM + CM

```python
data = pd.read_csv('singloblulos.csv')
```

```python
data= data.drop('CODIGO',axis=1)
```

```python
x = data.drop('SHOCK', axis=1)  # Ajusta 'etiqueta' al nombre de la columna de etiquetas
y = data['SHOCK']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify= y, random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
sampler = RandomUnderSampler(random_state=45)
X_train, y_train = sampler.fit_resample(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split
 
 
# Define un diccionario de combinaciones de hiperparámetros a probar
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'algorithm': ['SAMME', 'SAMME.R'],
    'random_state': [42, 123, 456]
}
 
# Crea una instancia del modelo AdaBoostClassifier con base en DecisionTree
adaboost_model = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(ccp_alpha= 0.0, criterion= "gini", max_depth= 20, max_features= "log2", min_impurity_decrease= 0.0, min_samples_leaf= 2, min_samples_split= 2, splitter= "best"))
 
# Realiza una búsqueda de hiperparámetros en conjunto (Grid Search)
grid_search = GridSearchCV(adaboost_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
 
# Obtiene los mejores hiperparámetros y la precisión correspondiente
best_params = grid_search.best_params_
best_accuracy = grid_search.best_score_
 
print("Mejor combinación de hiperparámetros:", best_params)
print("Precisión más alta encontrada:", best_accuracy)
 
# Entrena el modelo con los mejores hiperparámetros en el conjunto de entrenamiento completo
best_adaboost_model = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(ccp_alpha= 0.0, criterion= "gini", max_depth= 20, max_features= "log2", min_impurity_decrease= 0.0, min_samples_leaf= 2, min_samples_split= 2, splitter= "best"), **best_params)
best_adaboost_model.fit(X_train, y_train)
 
# Evalúa el modelo en el conjunto de prueba
y_pred = best_adaboost_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
 
# Calcula el valor AUC en el conjunto de prueba
y_prob = best_adaboost_model.decision_function(X_test)  # Obtén las puntuaciones de decisión para AUC
roc_auc = roc_auc_score(y_test, y_prob)
 
print("Precisión en el conjunto de prueba con los mejores hiperparámetros:", test_accuracy)
print("AUC en el conjunto de prueba con los mejores hiperparámetros:", roc_auc)
```
