```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('singloblulos.csv')
```

```python
data= data.drop('CODIGO',axis=1)
```

```python
x = data.drop('SHOCK', axis=1)  # Ajusta 'etiqueta' al nombre de la columna de etiquetas
y = data['SHOCK']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify= y, random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
sampler = RandomUnderSampler(random_state=45)
X_train, y_train = sampler.fit_resample(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

param_grid = {
    'n_estimators' : [10, 50, 100, 200],
    'criterion' : ["gini", "entropy"],
    'max_depth' : [10, 20, 30],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4],
    'max_features' : [None, "auto", "sqrt", "log2", 2, 3]
}

model = RandomForestClassifier()
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)  # X_train e y_train son tus datos de entrenamiento

best_params = grid_search.best_params_
best_accuracy = grid_search.best_score_

# Entrena el modelo con los mejores parmetros
best_model = RandomForestClassifier(**best_params)
best_model.fit(X_train, y_train)

# Realiza predicciones en el conjunto de prueba
y_pred = best_model.predict(X_test)

# Calcula la puntuacin ROC AUC
roc_auc = roc_auc_score(y_test, y_pred)

print("Mejor combinacin de hiperparmetros:", best_params)
print("Precisin ms alta encontrada:", best_accuracy)
print("Puntuacin ROC AUC:", roc_auc)
```
